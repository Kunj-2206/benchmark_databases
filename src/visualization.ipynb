```python
import matplotlib.pyplot as plt

# Database names
databases = ['Neo4j', 'MongoDB', 'DynamoDB', 'Google Bigtable']

# Time taken for each database (replace these values with actual times)
time_taken = [9.894041, 0.0270, 30.4528, 26.3094]  # Example times in seconds

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(databases, time_taken, color=['blue', 'green', 'orange', 'red'])

# Add labels and title
plt.xlabel('Database')
plt.ylabel('Time Taken (seconds)')
plt.title('Time Taken for Inserting 1000 Records in Different Databases')

# Display the plot
plt.show()
```
